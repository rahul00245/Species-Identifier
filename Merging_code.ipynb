```python
# Step 1: Import pandas
import pandas as pd

# Step 2: Load your datasets
# (Replace the filenames with your actual file paths or Google Drive links)
df1 = pd.read_csv('/content/dataset1.csv')
df2 = pd.read_csv('/content/dataset2.csv')

# Step 3: Merge the datasets
# You can merge based on a common column, for example 'id'
merged_df = pd.merge(df1, df2, on='id', how='inner')  # options: 'inner', 'left', 'right', 'outer'

# Step 4: Display the merged dataset
print(merged_df.head())

# Step 5: (Optional) Save the merged dataset
merged_df.to_csv('/content/merged_dataset.csv', index=False)
print("âœ… Merged dataset saved as merged_dataset.csv")
```
